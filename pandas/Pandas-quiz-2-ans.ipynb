# 1. 
다음 1월 수치(sales_1)와 2월 수치(sales_2)를 합쳐보자. 

```
np.random.seed(2020)
newspaper1 = ['경향신문신문', '국민일보신문', '동아일보신문', '문화일보신문', '서울신문신문', '세계일보신문','조선일보신문',  '중앙일보신문', '한겨레신문', '한국일보신문']
newspaper2 = ['경향신문신문', '국민일보신문', '동아일보신문', '문화일보신문', '서울신문신문', '세계일보신문']

sales1 = pd.Series(np.random.randint(1000, 9999, 10)*100, index=newspaper1)
sales2 = pd.Series(np.random.randint(1000, 9999, 6)*100, index=newspaper2)
print(sales1)
print(sales2)
```

# 4.
sales1 데이터에 score 라고 정의된 사용자 함수를 적용해서 각 신문사의 등급을 확인해보자.

```
def score(x):
    return 'A' if x > 500000 else 'B'
```
